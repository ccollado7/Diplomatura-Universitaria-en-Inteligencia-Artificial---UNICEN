#Preprocesamiento de texto

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.stem import PorterStemmer
from sklearn.feature_extraction import text

# Creamos el stemmer
stemmer = PorterStemmer()

# Construimos un TOKENIZER con la tokenización que provee CountVectorizer
tokenizer = TfidfVectorizer().build_tokenizer()

# Obtenemos las stopwords que provee SciKit Learn
stop_words = text.ENGLISH_STOP_WORDS.union('like')

# Definimos una función que aplica el stemming luego de tokenizar y remover stopwords
def  stem_tokenizer(doc):
    # Aplica la tokenización
    tokens = tokenizer(doc)
    # Retorna la lista de tokens luego de aplicar stemming si no es un stopword
    return list(stemmer.stem(w) for w in tokens if w not in stop_words)
```

#Entrenamos un Arbol de Decisión

```
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_validate
from sklearn.feature_extraction.text import TfidfVectorizer

# Crea el pipeline
pipe = make_pipeline(
     TfidfVectorizer(tokenizer=stem_tokenizer, ngram_range=(1,2), min_df=2),
     DecisionTreeClassifier()
 )

# Divide el df en X e y
X = df_quotes['Quote'].values
y = df_quotes['Author'].values


# Llama al metodo fit del pipeline, y entrena el clasificador con todos los datos disponibles
pipe = pipe.fit(X, y)
print(pipe['decisiontreeclassifier'].classes_)

# Realiza una predicción
print(pipe.predict(["I am Sherlock Holmes."]))

# Visualizamos el arbol resultante
fig, ax = plt.subplots(figsize=(50, 50))  # whatever size you want
plot_tree(pipe['decisiontreeclassifier'], feature_names=pipe['tfidfvectorizer'].get_feature_names(), class_names=pipe['decisiontreeclassifier'].classes_, max_depth=10, ax=ax, label='all')
plt.show()
```

```
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

# Genera las predicciones utilizando cross-validation
predicted = cross_val_predict(pipe, X, y, cv=5)

conf_mat = confusion_matrix(y, predicted)
print(conf_mat)

print(classification_report(y, predicted, target_names=df_quotes['Author'].unique()))
```
