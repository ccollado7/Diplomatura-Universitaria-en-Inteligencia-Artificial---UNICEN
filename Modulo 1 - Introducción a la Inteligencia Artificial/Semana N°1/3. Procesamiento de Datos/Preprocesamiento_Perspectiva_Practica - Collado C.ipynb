**Tratamiento de valores faltantes(Scikit-Learn)**

```
#Importo la libreria
from sklearn.impute import SimpleImputer 

#Concateno los dataframes X,y para tener todos los datos juntos
combined_data = pd.concat([X, y], axis=1)

#Elimino la columna con los targets (variable categorica)
dfdrop=combined_data.drop('class',axis=1)

#Instancio el imputador: Completo con la media para los NaN
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

#Realizo el fit del imputador
imputer = imputer.fit(dfdrop)

#Realizo la transformacion del dataframe por medio del imputador
imputed_data = pd.DataFrame(imputer.transform(dfdrop)) 

#Observo el dataframe imputado: Observar que el dataframe imputado no tiene columnas
print(imputed_data.head())

#En el proceso de imputacion se "pierden" el nombre de las columnas, por lo tanto necesito agregarlas nuevamente
imputed_data.columns=X.columns

#Concateno el dataframe imputado junto con el targer
combined_data_sci = pd.concat([imputed_data, y], axis=1)

#Observo el dataframe final
print(combined_data_sci)
```

**Manejo de texto y atributos categóricos (Scikit-Learn)**

**Agregamos atributos categéricos con valores al azar**

```
#Concateno los dataframes X,y para tener todos los datos juntos
combined_data = pd.concat([X, y], axis=1)

#Como no tengo features categoricos los agrego en forma artificial
combined_data['size'] = np.random.choice(['small','medium', 'large'], combined_data.shape[0]) #Atributo categorico ORDINAL (tiene cierto orden)
combined_data['color'] = np.random.choice(['red','green', 'blue'], combined_data.shape[0]) #Atributo categorico

#Observo como queda el dataframe
combined_data
```

**Atributos categóricos ordinales**

```
#Para los atributos categoricos ordinales puedo corresponder univocamente un numero que indique orden

#Diccionario que mapea las categorias ordinales con valores numericos
size_mapping = {
  'large': 3,
  'medium': 2,
  'small': 1}

#Modifico la columna categorica ordinal colocando los numeros correspondientes al mapeo
combined_data['size'] = combined_data['size'].map(size_mapping)
combined_data
```

**Atributos nominales: one hot encoding (Pandas)**

```
#Para las variables categoricas nominales utlizo one hot encoding con Pandas

combined_data_pan=pd.get_dummies(combined_data, prefix=['col'],columns=['color'])
combined_data_pan
```

**Atributos nominales: one hot encoding(Scikit-Learn)**

```
#Para las variables categoricas nominales utlizo one hot encoding con Scikit-Learn

#Importo las librerias
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Instancio la clase del transformador de columna
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [6])], remainder='passthrough')

#Realizo el fit y el transform del dataset
dataset = pd.DataFrame (columnTransformer.fit_transform(combined_data))

#Agrego los nombres de las columna
dataset.columns= pd.Index(['col_blue', 'col_green','col_red']).append(combined_data.columns.drop('color'))

#Observo como era el dataframe antes de la transformacion
print(combined_data)

#Observo como esta el dataframe transformado
dataset
```

**Codificación de etiquetas de clase**

```
#Recordemos que el target tenia nombre de las flores: El algoritmo necesita valores numericos para el entrenamiento, por lo tanto necesito transformar esos nombres en LABELS

#Importo el modulo para generar los labels
from sklearn.preprocessing import LabelEncoder

#Instancio el encoder
encoder = LabelEncoder()

#Selecciono la columna targee a la cual le voy a generar los labels
class_cat = combined_data["class"]

#Realizo fit y transform para la columna target
class_cat_encoded = encoder.fit_transform(class_cat)

#Reemplazo la columna de target original por la columna con los labels
combined_data["class"]=class_cat_encoded


print(class_cat_encoded)
print(encoder.classes_)

combined_data
```

**Transformación de datos: normalización min-max (Scikit-Learn)**

```
#Importo los modulos que voy a utilizar
from sklearn.preprocessing import MinMaxScaler #Escala a 0-1
from sklearn.compose import ColumnTransformer

#Instancio el transformador de columna
columnTransformer = ColumnTransformer([('num', MinMaxScaler(), [1])], remainder='passthrough') #Se instancia para transformar la primer columna por medio de [1]

#Realizo fit y transform sobre las columnas
dataset = pd.DataFrame(columnTransformer.fit_transform(combined_data))

#Incorporo los nombres a las columna
dataset.columns=combined_data.columns

#Observo el dataset transformado
dataset
```

**Transformación de datos de multiples atributos(Scikit-Learn)**

```
#Importo los modulos que utilizo
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#Aplico OneHotEncoder a una columna y MinMaxScaler a cuatro columna
#Defino una secuencia de operaciones por medio de una lista de tuplas
t = [('cat', OneHotEncoder(), [6]), ('num', MinMaxScaler(), [0,1,2,3])]

#Instancio el transformador e indico que utilice la secuencia de operaciones anteriores
columnTransformer = ColumnTransformer(transformers=t, remainder='passthrough')

#Realizo el fit y transform
dataset = pd.DataFrame(columnTransformer.fit_transform(combined_data))

#Observo el dataset transformado
dataset
```

**Transformación  de datos de multiples atributos en pipeline(Scikit-Learn)**

```
#Importo los modulos que voy a utilizar

from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline #Permite contruir pipeline

# Creo un pipeline para los features numericos: Realizo imputacion para los NaN con la media y luego escalo los valores
numeric_features = ['sepallength',	'sepalwidth',	'petallength',	'petalwidth']
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())])

#Creo un pipeline para los features categoricos: Realizo imputacion para los NaN rellenando con 'missing' y luego realizo OneHotEncoder
categorical_features = ['color']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder())])

#Utilizando los pipelines para features numericos y categoricos realiza una transformacion por columna aplicandolos

#Instancio el transformador por columna
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)],remainder='passthrough')

#Realizo el fit y transform
dataset = pd.DataFrame(preprocessor.fit_transform(combined_data))

#Observo el dataframe transformado
dataset

```
